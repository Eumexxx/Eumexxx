From, temperature plot we can see there is a outliers with temperature value being 400 degree celsius, which is not a possible value for earth atmospheric temperature (Screenshot attached) .Also, there seems to be 24 missing values in Temperature Column(can be seen in data wrangling step).image.png

**Handling outliers and Missing value**

Outliers value is replaced by average value of one value ahead and one value before of the outlier. Similarly, missing values are imputed with past 7 days average value for each missing hour.

````python
# Handling Outliers
outlier_value = 400
outlier_index = df1[df1['Temperature'] == outlier_value].index

# Check if any elements in outlier_index are in df2.index
if any(idx in df1.index for idx in outlier_index):
    for idx in outlier_index:
        idx_timestamp = pd.Timestamp(idx)  # Convert idx to Timestamp
        prev_timestamp = idx_timestamp - pd.Timedelta(hours=1)
        next_timestamp = idx_timestamp + pd.Timedelta(hours=1)

        # Check if the Timestamps exist in the index before accessing them
        if prev_timestamp in df1.index and next_timestamp in df1.index:
            surrounding_temps = df1.loc[[prev_timestamp, next_timestamp], 'Temperature']
            avg_temp = surrounding_temps.mean()
            df1.at[idx, 'Temperature'] = avg_temp

# Handling Missing Values
for idx in df1[df1['Temperature'].isna()].index:
    # Get the hour of the missing data
    missing_hour = idx.hour

    # Define the period of the past 7 days
    start_time = idx - pd.Timedelta(days=7)
    end_time = idx

    # Filter data from the past 7 days and the same hour, then compute the average temperature
    avg_temp = df1[(df1.index.hour == missing_hour) & (df1.index >= start_time) & (df1.index < end_time)]['Temperature'].mean()

    # Impute the missing value with the computed average
    df1.at[idx, 'Temperature'] = avg_temp
````

# 4. Preprocessing and Training



**Feature Engineering**

Here new feature 'Corrt' which is the correlation value between GHI and Temperature is added.The intension for doing so is to help model better understand the relationship between input variables  (GHI & Temperature).

Steps taken

correlation is typically calculated over a range of data points, not for individual data points.So rolling or moving correlation over a window of rows is calculated. Rolling window of 5 consecutive hourly datapoint is considered for this analysis.

* Iterating Through the DataFrame with a Rolling Window:

The code iterates through df1 using a rolling window approach. This is done by looping from the start of the DataFrame (start = 0) to len(df1) - 5. The -5 ensures that the window does not extend beyond the end of the DataFrame.
For each iteration, a subset of the DataFrame (window) is selected, spanning 5 rows starting from the current index (start). This subset represents the current rolling window of data.

* Calculating Correlation:

Within each rolling window, the Pearson correlation coefficient between the 'GHI' and 'Temperature' columns is calculated. This coefficient is a statistical measure of the linear relationship between two variables, ranging from -1 to 1.
The calculated correlation (corr) for each window is assigned to the corresponding position in the correlation_values list. Specifically, it's assigned to the index start + 5, which refers to the last row of the current window.

* Adding the 'Corrt' Column to the DataFrame:

The correlation_values list, now containing the calculated correlations for each rolling window, is used to create a new column in df1 named 'Corrt'.
Initially, rows without a calculated correlation (like the first few rows) have None as their value in 'Corrt'. These are then filled with 0 using the fillna(0) method. This step replaces all None values with 0, indicating no correlation.


````python
import pandas as pd

correlation_values = [None] * len(df1)  # Initialize with None

# Iterate through the DataFrame using a rolling window
for start in range(len(df1) -5):
    window = df1.iloc[start:start + 5]
    corr = window['GHI'].corr(window['Temperature'])  #Pearson Correlation between GHI  and Temperature
    correlation_values[start + 5] = corr  # Assign to the last row of the window

df1['Corrt'] = correlation_values
df1['Corrt'] = df1['Corrt'].fillna(0)
df1.tail()



````

# 5. Modeling

**Univariate ARIMA Model**

In ARIMA, the term AR, I, and MA represents the following responsibilities. AR (Auto-Regressive) model is responsible for understanding the correlation between the observed value and its lagged value. I (Integrated) model is responsible to make the best fit of data by substitution of data with the differenced term (to make the data stationary) . MA (Moving Average) model is responsible for understanding the correlation between an observed value and a residual error in the lagged data. Each of the above terms is associated with the parameters represented as (p, d, q), in which:

•	p: corresponds to the quantity of lag observation in a model.

•	d: corresponds to the number of times the observed value is different from its lagged value.

•	q: corresponds to the order of lagged prediction error.


image.png

````python

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA


# Fit ARIMA model on training data with random p,d,q (3, 0, 5) value for simplicity ( determining these value based on lowest AIC(Akaike Information Criterion) value will be better, If you have higher computational power. The reason for training with random value is to reduce computational time )
model = ARIMA(train['GHI'], order=(3,0,5))
fit_model = model.fit()

# Predict the last 72 hours(of test data)
forecast = fit_model.forecast(steps=72)

# For better plotting
forecast.index = test.index

# Plotting the results
plt.figure(figsize=(14, 7))
plt.plot(test.index, test['GHI'], label="Actual GHI", color='red')
plt.plot(forecast.index, forecast, label="Predicted GHI", color='blue')
plt.title("Actual vs Predicted GHI for the Last 72 hours")
plt.xlabel("DateTime")
plt.ylabel("GHI")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
#It takes approx. 3-5 min for complete running of this code
````

**Random Forest Model (Multivariate Model)**

Random Forrest regression is a supervised learning approach for regression that uses the bagging ensemble learning method. The random forest model is introduced to improve the performance of the machine learning algorithm "decision tree". The issue with decision tree algorithms is that they tend to overfit with smaller datasets. The RF approach resolves this problem by picking dataset features randomly.

````python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor


# Select the columns for training and testing
features = ['Temperature', 'DHI', 'Solar Zenith Angle', 'Surface Albedo', 'Pressure', 'Wind Direction', 'Wind Speed','Corrt'] #DNI varible is not considered for training because of very high correlation with GHI (check correlation plot).
target = 'GHI'

#splitting the data
train = df1[-72:]
test = df1[-72:]

# Split features and target for training and testing
X_train, y_train = train[features], train[target]
X_test, y_test = test[features], test[target]

# Create the RF regressor model and fit it to the training data
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Make predictions on the test set
predictions = rf.predict(X_test)

# Plot the results for the last 72 hours
plt.figure(figsize=(14, 7))
plt.plot(test.index, y_test, label="Actual GHI", color='red')
plt.plot(test.index, predictions, label="Predicted GHI", color='blue')
plt.title("Actual vs Predicted GHI for the Last 72 hours")
plt.xlabel("Last 72 hours")
plt.ylabel("GHI")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


#This code takes around 2 min to make a complete run
````
